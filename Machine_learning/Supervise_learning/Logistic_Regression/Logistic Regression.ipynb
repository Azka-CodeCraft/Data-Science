```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ir=pd.read_csv("Iris.csv")
ir
```

```python
ir.drop("Id",axis=1,inplace=True)
```

```python
ir
```

```python
uv=ir['Species'].unique()
uv
```

```python
#ir=ir[ir['Species']=='Iris-setosa']
```

```python
#ir
```

```python
ir=ir[ir['Species'] != 'Iris-setosa']
```

```python
ir
```

```python
ir['Species']=ir['Species'].map({'Iris-versicolor':0,'Iris-virginica':1})
ir
```

```python
ir.tail()
```

```python
ir
```

```python
X = ir.iloc[:, :-1]
X
```

```python
y = ir.iloc[:, -1]
y
```

```python
cross validation train data unseen data ki tarining
```

```python
from sklearn.model_selection import train_test_split
train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size = 0.30, random_state=42)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lm= LogisticRegression()
```

```python
lm
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
c_regressor=GridSearchCV(lm,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
c_regressor
```

```python
c_regressor.fit(X_train,y_train)
```

```python
c_regressor.best_params_
```

```python
c_regressor.best_score_
```

```python
pred=c_regressor.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy=accuracy_score(y_test,pred)
accuracy
```

```python
ac=classification_report(y_test,pred)
print(ac)
```
